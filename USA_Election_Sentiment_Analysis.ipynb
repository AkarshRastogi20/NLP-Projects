```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
import plotly.graph_objects as go
import plotly.express as px
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
trump=pd.read_csv('/content/Trumpall2.csv')
biden=pd.read_csv('/content/Bidenall2.csv')
```

```python
# create function for checking +ve and -ve comment

def find_polarity(review):
  return TextBlob(review).sentiment.polarity
```

```python
find_polarity('I hate you') # output shows that it is -ve comment
```

```python
trump['Sentiment Polarity']=trump['text'].apply(find_polarity)
biden['Sentiment Polarity']=biden['text'].apply(find_polarity)
```

```python
trump
```

```python
biden
```

```python
# convert polarity number as +ve and -ve and neutral

trump['Expression Label']=np.where(trump['Sentiment Polarity']>0,'Positive','Negative')
trump['Expression Label'][trump['Sentiment Polarity']==0]='Neutral'
```

```python
biden['Expression Label']=np.where(biden['Sentiment Polarity']>0,'Positive','Negative')
biden['Expression Label'][biden['Sentiment Polarity']==0]='Neutral'
```

```python
trump
```

```python
biden
```

**drop all the neutral polarity tweets**

```python
# drop all the neutral polarity tweets

n_trump=trump[trump['Sentiment Polarity']==0]
n_biden=biden[biden['Sentiment Polarity']==0]
print(n_trump.shape)
print(n_biden.shape)
```

```python
cond1=trump['Sentiment Polarity'].isin(n_trump['Sentiment Polarity'])
trump.drop(trump[cond1].index, inplace=True)
print(trump.shape)
cond2=biden['Sentiment Polarity'].isin(n_biden['Sentiment Polarity'])
biden.drop(biden[cond2].index, inplace=True)
print(biden.shape)
```

```python
trump
```

```python
biden
```

**Balance the dataset**

```python
np.random.seed(10)
remove_n=324
drop_indices=np.random.choice(trump.index, remove_n, replace=False)
df_subset_trump=trump.drop(drop_indices)
print(df_subset_trump.shape)
```

```python
np.random.seed(10)
remove_n=31
drop_indices=np.random.choice(biden.index, remove_n, replace=False)
df_subset_biden=biden.drop(drop_indices)
print(df_subset_biden.shape)
```

**Predict the USA Elections**

```python
count_1=df_subset_trump.groupby('Expression Label').count()
```

```python
neg_per1=(count_1['Sentiment Polarity'][0]/1000)*10
pos_per1=(count_1['Sentiment Polarity'][1]/1000)*100
print(pos_per1)
print(neg_per1) # showing the % of +ve and -ve comment
```

```python
count_2=df_subset_biden.groupby('Expression Label').count()
neg_per2=(count_2['Sentiment Polarity'][0]/1000)*100
pos_per2=(count_2['Sentiment Polarity'][1]/1000)*100
print(pos_per2)
print(neg_per2)
```

```python
# Plot the diagram using plotly

politicians=['Biden','Trump']

pos_list=[pos_per2,pos_per1]
neg_list=[neg_per2,neg_per1]

fig=go.Figure(data=[
    go.Bar(name='Positive',x=politicians,y=pos_list),
    go.Bar(name='Negative',x=politicians,y=neg_list)
])

fig
```
